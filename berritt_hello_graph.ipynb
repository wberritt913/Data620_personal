```python
# import packages
import networkx as nx
import matplotlib.pyplot as plt

# create a graph object
g = nx.Graph()

# node list
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix",
    "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose",
    "Austin", "Jacksonville", "Fort Worth", "Columbus", "Charlotte",
    "San Francisco", "Indianapolis", "Seattle", "Denver", "Washington D.C."]

# add nodes to graph
g.add_nodes_from(cities)

# edge list
edges = [("New York", "Los Angeles"), ("New York", "Chicago"), ("New York", "Houston"),
    ("Los Angeles", "Phoenix"), ("Los Angeles", "San Diego"), ("Los Angeles", "San Francisco"),
    ("Chicago", "Dallas"), ("Chicago", "Philadelphia"), ("Chicago", "Indianapolis"),
    ("Houston", "San Antonio"), ("Houston", "Dallas"), ("Houston", "Austin"),
    ("Phoenix", "San Diego"), ("Phoenix", "San Antonio"),
    ("Philadelphia", "Charlotte"), ("Philadelphia", "Washington D.C."),
    ("San Antonio", "Dallas"), ("San Antonio", "Austin"),
    ("San Diego", "San Francisco"), ("San Diego", "Denver"),
    ("Dallas", "Austin"), ("Dallas", "Fort Worth"),
    ("San Jose", "San Francisco"), ("San Jose", "Seattle"),
    ("Austin", "Jacksonville"), ("Austin", "Columbus"),
    ("Jacksonville", "Charlotte"), ("Jacksonville", "Washington D.C."),
    ("Fort Worth", "Denver"), ("Fort Worth", "Seattle"),
    ("Columbus", "Indianapolis"), ("Columbus", "Charlotte"),
    ("Charlotte", "Washington D.C."), ("Charlotte", "Indianapolis"),
    ("San Francisco", "Seattle"), ("San Francisco", "Denver"),
    ("Indianapolis", "Denver"), ("Indianapolis", "Washington D.C."),
    ("Seattle", "Denver"), ("Seattle", "San Francisco"),
    ("Denver", "Washington D.C."), ("Denver", "San Francisco")]

# add edges
g.add_edges_from(edges)

# spring graph
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(g, k=0.3)
nx.draw(g, pos, with_labels=True, node_size=3000, font_size=10, font_weight='bold', edge_color='gray', linewidths=1, font_color='black')
plt.title("Network of Cities and Direct Flights", size=15)
plt.show()

# circle graph
plt.figure(figsize=(12, 10))
nx.draw_circular(g, with_labels=True, node_size=3000, font_size=10, font_weight='bold', edge_color='gray', linewidths=1, font_color='black')
plt.title("Network of Cities and Direct Flights (Circular Layout)", size=15)
plt.show()
```
